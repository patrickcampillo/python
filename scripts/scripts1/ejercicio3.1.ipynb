```python
def calcul():
    print("\nEsta calculadora calcula el área o volumén de algunos póligonos, escoge el que quieras calcular.\n ")
    print("1.Área de un triángulo.")
    print("2.Área de un cuadrado.")
    print("3.Área de un petágono.")
    print("4.Área de un hexágono.")
    print("5.Volumen de una esfera.")
    print("6.Volumen de una pirámide.")
    print("7.Volumen de un cilindro.")
    print("8.Volumen de un icosaedro.")
    print("9.Vuelve a mostrar el menú.")
    print("10.Salir")
    
calcul()
ope=float(input("\nIndica la operación a realizar: "))

while (ope > 0 and ope < 10):
    
    if ope == 1:
        b=float(input("Introduce la longitud de la base: "))
        h=float(input("Introduce la altura: "))
        resul=b*h/2
        print("El área del triángulo es", resul)
        
    elif ope == 2:
        l=float(input("Introduce la longitud de los lados: "))
        resul=l**2
        print("El área del cuadrado es", resul)
        
    elif ope == 3:
        l=float(input("Introduce la longitud de los lados: "))
        import math
        num=36*math.pi/180
        tan=math.tan(num)
        resul=(5*l**2)/(4*tan)
        print("El área del pentágono es",resul)
        
    elif ope == 4:
        l=float(input("Introduce la longitud de los lados: "))
        import math
        num=30*math.pi/180
        tan=math.tan(num)
        resul=(3*l**2)/(2*tan)
        print("El área del hexágono es",resul)
        
    elif ope == 5:
        r=float(input("Introduce la longitud del radio: "))
        from math import pi
        resul=4/3*pi*r**3
        print("El volumen de la esfera es",resul)
        
    elif ope == 6:
        h=float(input("Introduce la altura: "))
        l=float(input("Introduce la longitud del lado de la base: "))
        resul=(h*l**2)/3
        print("El volumen de la piramide es",resul)
        
    elif ope == 7:
        h=float(input("Introduce la altura: "))
        r=float(input("Introduce la longitud del radio: "))
        from math import pi
        resul=h*pi*r**2
        print("El volumen del cilindro es",resul)
        
    elif ope == 8:
        arista=float(input("Introduce la longitud de la arista: "))
        import math
        raiz=math.sqrt(5)
        resul=5/12*(3+raiz)*arista**3
        print("El volumen del icosaedro es",resul)
    
    elif ope == 9:
        calcul()

    ope=float(input("\nIndica la operación a realizar: "))
```
