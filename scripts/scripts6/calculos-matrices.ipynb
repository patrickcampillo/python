```python
def read_matrix(a,b):    #función que genera una matriz con valores entre -10 y 20 de la longitud  y tamaño indicado
    return np.random.randint(-10, 20,(a,b))
```

```python
def show_matrix(a):  #Función que muestra las matrices ordenadas por columna
    print()
    for i in a:
        for valor in i:
            print("\t", valor, end=" ")   #El valor \t añade una tabulación a la salida, y se le indica con el end que no salte de línea
        print("\n")
```

```python
def sum_matrix(a,b,c):  #Función que realiza la suma de dos matrices, para sumar matrices se han de sumar el valor que contiene cada posición
                        #de la primera matriz con el valor de la misma posición de la segunda
    
    for i in range(len(a)):     #Bucle que controla cada vector de la matriz
        
        for s in range(len(a[i])): #Bucle que controla cada valor del vector y almacena la suma de los valores en una nueva matriz
            c[i][s] = a[i][s] + b[i][s]
                
    return c.astype(int)    #astype devuelve la variable en el tipo indicado
```

```python
def rest_matrix(a,b,c):  #Función que realiza la resta de dos matrices, para restar matrices se han de restar el valor que contiene cada posición
                         #de la primera matriz con el valor de la misma posición de la segunda, prestando atención al símbolo.
    
    for i in range(len(a)):     #Bucle que controla cada vector de la matriz
        
        for s in range(len(a)): #Bucle que controla cada valor del vector y almacena la resta de los valores en una nueva matriz
            c[i][s] = a[i][s] - b[i][s]
                
    return c.astype(int)    #astype devuelve la variable en el tipo indicado
```

```python
def multi_matrix(a,b,c):  #Función que realiza la multiplicación de dos matrices, es necesario que el número de columnas
                          #de la primera matriz sea igual al número de filas de la segunda. Y sabemos que la matriz resultante
                          #tendrá una dimensión equivalente al número de filas de la primera matriz y el número de columnas de la segunda
    
    #Necesitamos 3 bucles para controlar las filas y columnas de cada matriz, sabiendo que las filas de la
    #segunda matriz tiene que ser igual a las columnas de la primera par poderse multiplicar.
    
    for i in range(len(a)):  #Bucle que controla las filas de la primera matriz
        for s in range(dimension(b)[1]):  #Buvle que controla las columnas de la segunda matriz
            for z in range(len(b)):   #Bucle que controla las filas de la segunda matriz
                c[i][s] += a[i][z] * b[z][s]    
                
    return c.astype(int)    #astype devuelve la variable en el tipo indicado
```

```python
def comp_multi(a,b): #Función que comprueba si se pueden multiplicar las matrices, comprobando si el número de columnas 
                     #de la primera matriz es igual al de las filas de la segunda.
    
    if ((dimension(a)[1] == dimension(b)[0])):
        return True
    
    else:
        print("\nEl número de filas de la segunda matriz no es igual al de las columnas de la primera.")
        return False
```

```python
def dimension(a):   #Función que devuelve la dimensión de una matriz introducida
    return np.shape(a)
```

```python
def elegir_divisor(a,b): #Función que establece que matriz será divisora, 
                         #si alguna de las dos cumple el requisito de ser cuadrada
        
    dimension1 = dimension(a)
    dimension2 = dimension(b)
    
    if dimension2[0] == dimension2[1]:
        return np.linalg.inv(b),a
    
    elif dimension1[0] == dimension1[1]:
        print("\nYa que la segunda matriz no es cuadrada establecemos la primera como divisora.\n")
        return np.linalg.inv(a),b
    
    else:
        print("\nNo se puede dividir, ya que ninguna de las matrices es cuadrada y no se puede obtener su inversa.\n")
        return False
```
